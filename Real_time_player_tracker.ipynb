```python
from ultralytics import YOLO

model = YOLO("best.pt")
```

```python
import cv2

# Open the uploaded video
video_path = "15sec_input_720p.mp4"  # Replace if your file name is different
cap = cv2.VideoCapture(video_path)

# Get video info
width = int(cap.get(3))
height = int(cap.get(4))
fps = int(cap.get(5))

# Define output video writer
out = cv2.VideoWriter("output_detected.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)

    # Annotate frame with results
    annotated = results[0].plot()  # Draw bounding boxes and labels

    # Write frame to output video
    out.write(annotated)

cap.release()
out.release()
```
